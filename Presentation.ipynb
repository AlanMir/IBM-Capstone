## 1.Introduction
### 1.1 Background
#### Los Mochis is a small coastal city in northwest Mexico. Due to its location, seafood cuisine is highly popular, and new business are being created every year. However, many of these are not succesful, and they do not atract enough customers to grow. One of the most important factors for a restaurant survival is its location. A good selection of this feature can give restaurants a competitive advantage over their competitors and hence leverage the construction of a recognized and succesful brand.

### 1.2 Research question
#### What is the most suitable location for a new seafood restaurant in Los Mochis?
